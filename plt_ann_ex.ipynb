```python
import graphviz
import numpy as np
```

```python
x_n = np.arange(0,4,1)
z0_n = np.arange(0,8,1)
z1_n = np.arange(0,8,1)
z2_n = np.arange(0,8,1)
y_n = np.arange(0,2,1)


g = graphviz.Digraph(comment='COMMENT')
g.attr(rankdir='LR', size='8.5', engine='neato', ratio = '1')

# x
with g.subgraph(name='cluster_0') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='input')
    
    for i in x_n:
        x_label = f'x{i}'
        c.node(x_label,x_label)
            
        
# y
with g.subgraph(name='cluster_1') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='output layer')
    
    for i in y_n:
        y_label = f'y{i}'
        c.node(y_label,y_label)
        
        for j in x_n:
            x_label = f'x{j}'
            c.edge(x_label, y_label)
            
    
g.render('linMdl', 'pdf')
g
```

```python
x_n = np.arange(0,4,1)
z0_n = np.arange(0,8,1)
z1_n = np.arange(0,8,1)
z2_n = np.arange(0,8,1)
y_n = np.arange(0,2,1)


g = graphviz.Digraph(comment='COMMENT')
g.attr(rankdir='LR', size='8.5', engine='neato', ratio = '0.5')

# x
with g.subgraph(name='cluster_0') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='input')
    
    for i in x_n:
        x_label = f'x{i}'
        c.node(x_label,x_label)
        
# z0
with g.subgraph(name='cluster_1') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='first layer')
    
    for i in z0_n:
        z0_label = f'z0{i}'
        c.node(z0_label,z0_label)
        
        for j in x_n:
            x_label = f'x{j}'
            c.edge(x_label, z0_label)
            
# z1
with g.subgraph(name='cluster_2') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='second layer')
    
    for i in z1_n:
        z1_label = f'z1{i}'
        c.node(z1_label,z1_label)
        
        for j in z0_n:
            z0_label = f'z0{j}'
            c.edge(z0_label, z1_label)
            
# z2
with g.subgraph(name='cluster_3') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='third layer')
    
    for i in z2_n:
        z2_label = f'z2{i}'
        c.node(z2_label,z2_label)
        
        for j in z1_n:
            z1_label = f'z1{j}'
            c.edge(z1_label, z2_label, len='10')
            
            
            
# y
with g.subgraph(name='cluster_4') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='output layer')
    
    for i in y_n:
        y_label = f'y{i}'
        c.node(y_label,y_label)
        
        for j in z2_n:
            z2_label = f'z2{j}'
            c.edge(z2_label, y_label, len='10')
            
    
g.render('annMdl', 'pdf')
g
```

```python
from graphviz import Digraph

g = Digraph('G', filename='cluster.gv')

# NOTE: the subgraph name needs to begin with 'cluster' (all lowercase)
#       so that Graphviz recognizes it as a special cluster subgraph

with g.subgraph(name='cluster_0') as c:
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.edges([('a0', 'a1'), ('a1', 'a2'), ('a2', 'a3')])
    c.attr(label='process #1')

with g.subgraph(name='cluster_1') as c:
    c.attr(color='blue')
    c.node_attr['style'] = 'filled'
    c.edges([('b0', 'b1'), ('b1', 'b2'), ('b2', 'b3')])
    c.attr(label='process #2')

g.edge('start', 'a0')
g.edge('start', 'b0')
g.edge('a1', 'b3')
g.edge('b2', 'a3')
g.edge('a3', 'a0')
g.edge('a3', 'end')
g.edge('b3', 'end')

g.node('start', shape='Mdiamond')
g.node('end', shape='Msquare')

g.render('my_model_G', 'pdf')
g
```
