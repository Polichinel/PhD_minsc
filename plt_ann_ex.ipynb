```python
import graphviz
import numpy as np
```

```python
x_n = np.arange(0,4,1)
z0_n = np.arange(0,8,1)
z1_n = np.arange(0,8,1)
z2_n = np.arange(0,8,1)
y_n = np.arange(0,2,1)


g = graphviz.Digraph(comment='COMMENT')
g.attr(rankdir='LR', size='8.5', engine='neato', ratio = '1')

# x
with g.subgraph(name='cluster_0') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='input')
    
    for i in x_n:
        x_label = f'x{i}'
        c.node(x_label,x_label)
            
        
# y
with g.subgraph(name='cluster_1') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='output')
    
    for i in y_n:
        y_label = f'y{i}'
        c.node(y_label,y_label)
        
        for j in x_n:
            x_label = f'x{j}'
            c.edge(x_label, y_label)
            
    
g.render('linMdl', 'pdf')
g
```

```python
x_n = np.arange(0,4,1)
z0_n = np.arange(0,6,1)
z1_n = np.arange(0,6,1)
z2_n = np.arange(0,6,1)
y_n = np.arange(0,2,1)


g = graphviz.Digraph(comment='COMMENT')
g.attr(rankdir='LR', size='8.5', engine='neato', ratio = '0.5')

# x
with g.subgraph(name='cluster_0') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='input')
    
    for i in x_n:
        x_label = f'x{i}'
        c.node(x_label,x_label)
        
# z0
with g.subgraph(name='cluster_1') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='first layer')
    
    for i in z0_n:
        z0_label = f'z0{i}'
        c.node(z0_label,z0_label)
        
        for j in x_n:
            x_label = f'x{j}'
            c.edge(x_label, z0_label)
            
# z1
with g.subgraph(name='cluster_2') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='second layer')
    
    for i in z1_n:
        z1_label = f'z1{i}'
        c.node(z1_label,z1_label)
        
        for j in z0_n:
            z0_label = f'z0{j}'
            c.edge(z0_label, z1_label)
            
# z2
with g.subgraph(name='cluster_3') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='third layer')
    
    for i in z2_n:
        z2_label = f'z2{i}'
        c.node(z2_label,z2_label)
        
        for j in z1_n:
            z1_label = f'z1{j}'
            c.edge(z1_label, z2_label, len='10')
            
            
            
# y
with g.subgraph(name='cluster_4') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='output')
    
    for i in y_n:
        y_label = f'y{i}'
        c.node(y_label,y_label)
        
        for j in z2_n:
            z2_label = f'z2{j}'
            c.edge(z2_label, y_label, len='10')
            
    
g.render('annMdl', 'pdf')
g
```

```python
i0_n = np.arange(0,6,1)
i1_n = np.arange(0,6,1)
i2_n = np.arange(0,6,1)


g = graphviz.Digraph(comment='COMMENT')
g.attr(rankdir='LR', size='8.5', engine='neato', ratio = '1.5')

# i0
with g.subgraph(name='clusterA') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='Raw bin')
    
    for i in i0_n:
        i0_label = f'image 0{i}'
        c.node(i0_label,i0_label)

# i1
with g.subgraph(name='clusterB') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='Edit bin')
    
    for i in i1_n:
        i1_label = f'image 1{i}'
        c.node(i1_label,i1_label)

            
# i2
with g.subgraph(name='cluster_3') as c:
    
    c.attr(style='filled', color='lightgrey')
    c.node_attr.update(style='filled', color='white')
    c.attr(label='Published bin')
    
    for i in i2_n:
        i2_label = f'image 2{i}'
        c.node(i2_label,i2_label)
        
# g.edge('clusterA', 'clusterB')        
g.edge(i0_label, i1_label, ltail='cluster_0', lhead='cluster_1')
g.edge(i1_label, i2_label, ltail='cluster_1', lhead='cluster_2')


g.render('bins', 'pdf')
g
```
