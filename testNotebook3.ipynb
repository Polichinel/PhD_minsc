```python
import os
import cv2
from iptcinfo3 import IPTCInfo
from PIL import Image
import re
```

```python
# Dir with raw data
DATADIR_RAW = 'C:/Users/xpn381/Documents/data/wetransfer-d44dd0/Simon_Tests_12092019/raw'

# Dir with edit data - containing valid IPTC
DATADIR_EDIT = 'C:/Users/xpn381/Documents/data/wetransfer-d44dd0/Simon_Tests_12092019/edit'

# Dir for processed raws - now containing valid IPTC
DATADIR_NEW = 'C:/Users/xpn381/Documents/data/wetransfer-d44dd0/Simon_Tests_12092019/raw_w_dates_n_info'
```

```python
def insert_IPTC(DATADIR_RAW,DATADIR_NEW):
    
    # list to hold all edit dates reiterated over
    list_of_dates = []
    
    for edit_img in os.listdir(DATADIR_EDIT):
        edit_image_path = os.path.join(DATADIR_EDIT, edit_img)
#         print(edit_image_path)

        edit_info = IPTCInfo(edit_image_path, force=True)
        edit_date = str(edit_info['date created'])[2:10] # assumning this holds for all
        
        # If you already checked the date don't bother
        if edit_date not in list_of_dates:
            list_of_dates.append(edit_date) #appending all reiterated over
            
            # reiterating all over raw
            for raw_img in os.listdir(DATADIR_RAW):
                
                # force open the IPTC
                raw_image_path = os.path.join(DATADIR_RAW, raw_img)
                raw_info = IPTCInfo(raw_image_path, force=True)
                
                # Open the image and access the meta data. Use regex to extract date:
                raw_image = Image.open(raw_image_path) 
                txt = str(raw_image.info['exif']) # exif are mostly JPEG
                r  = '[\d]{4}:[\d]{2}:[\d]{2}' # format yyyy:mm:dd
                raw_date = re.search(r, txt).group().replace(':','')
                
                # Check if raw date matches edit data. If TRUE we assume same location
                if edit_date == raw_date:# This is the condition that migt have to change...

                    raw_info['date created'] = raw_date # could come before..
                    
                    raw_info['city'] = edit_info['city']
                    raw_info['country/primary location code'] = edit_info['country/primary location code']
                    raw_info['country/primary location name'] = edit_info['country/primary location name']
                    raw_info['copyright notice'] = edit_info['copyright notice']
                    
                    # Create the new dir DATADIR_NEW if it does not already exist
                    if not os.path.exists(DATADIR_NEW):
                        os.makedirs(DATADIR_NEW)
        
                    # Create the new path for the image
                    new_path = os.path.join(DATADIR_NEW, 'IPTC_' + raw_img) #better names...
        
                    # If that image is already in the path, delete it.
                    if os.path.exists(DATADIR_NEW):
                        try:
                            os.remove(new_path)
                        except:
                            pass
        
                    # Save new image.
                    raw_info.save_as(new_path)
                    print(f'saved: {new_path}')
                    
        
insert_IPTC(DATADIR_RAW,DATADIR_NEW)
```
